```python
import pandas as pd
%matplotlib
```

```python
heart_info_df = pd.read_csv("file.csv")
```

```python
heart_info_df
```

```python
# BPM * 60 minutes * 24 hours * 365 days = beats per year
beats_per_year = heart_info_df['Resting Heart Rate (BPM)'] * 60 * 24 * 365
```

```python
heart_info_df['Beats Per Year'] = beats_per_year
```

```python
heart_info_df
```

```python
# beats per year / longevity (years) = total heartbeats
lifetime_heartbeats = heart_info_df['Beats Per Year']/heart_info_df['Longevity (Years)']
```

```python
heart_info_df['Lifetime Heartbeats'] = lifetime_heartbeats
```

```python
heart_info_df
```

```python
heart_info_df['Lifetime Heartbeats'].plot(kind='bar')
```

```python
heart_info_df['Lifetime Heartbeats'].sort_values(ascending=False).plot(kind='bar')
```
